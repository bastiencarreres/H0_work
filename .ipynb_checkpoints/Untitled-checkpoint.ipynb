```python
#DEFINITION DES MESURES EXTERNES
av=0.71273
sigma_av=0.00176

mu_N4258 = 29.387
sigma_mu_N4258 = 0.0568
```

```python
sn_host=[b'M101', b'N1015', b'N1309', b'N1365', 
                      b'N1448',b'N2442', b'N3021', 
                      b'N3370', b'N3447', b'N3972', 
                      b'N3982', b'N4038', b'N4424', 
                      b'N4536', b'N4639', b'N5584', 
                      b'N5917', b'N7250', b'U9391']
```

## Fit with one anchor : NGC4258 

```python
def chi2(v):
    n = len(sn_host)
    mu_mu4258=v[:n]
    zp_w4258=v[n]
    b_w = v[n+1]
    m_v4258 = v[n+2]
    Z_w = v[n+3]
    chi2 = 0
    for host,i in zip(sn_host,np.arange(n)):
        selec_sn = data_sn[data_sn['Field']==host]
        m_sn_mod = m_v4258 + mu_mu4258[i]
        m_sn = selec_sn['mag_5a'] - 5*av
        cov_sn = pw(selec_sn['sigma_mag_5a'],2)-25*pw(sigma_av,2)
        chi2 += pw(m_sn-m_sn_mod,2)/cov_sn
        
        selec_ceph = data_cephe[data_cephe['Field']== host]
        C_ceph_inv=np.linalg.inv(np.diag(pw(selec_ceph['sigma_tot'],2)))
        m_ceph_mod = mu_mu4258[i] + zp_w4258 + b_w*np.log10(selec_ceph['Per']) + Z_w*(selec_ceph['[O/H]']-12)
        m_ceph = selec_ceph['F160W']-R*selec_ceph['F555W-F814W']
        chi2 += (m_ceph-m_ceph_mod).T @ C_ceph_inv @ (m_ceph-m_ceph_mod)
        
    ceph_4258 = data_cephe[data_cephe['Field']==b'N4258']
    m_ceph_mod = zp_w4258 + b_w*np.log10(ceph_4258['Per']) + Z_w*(ceph_4258['[O/H]']-12)
    m_ceph = ceph_4258['F160W']-R*ceph_4258['F555W-F814W']
    C_ceph_inv=np.linalg.inv(np.diag(pw(ceph_4258['sigma_tot'],2)))
    chi2 += (m_ceph-m_ceph_mod).T @ C_ceph_inv @ (m_ceph-m_ceph_mod)
    
    return chi2
```

```python
selec_sn = data_sn[data_sn['Field']==b'M101']
```

```python
#x0 =np.concatenate((np.array([2]*len(sn_host)),np.array([0.,-3.,20,-0.3])))
x0= np.array([-0.17355017,  3.15163952,  3.12488155,  1.93346946,  1.96991491,
        2.12292654,  3.04878019,  2.7036323 ,  2.55459972,  2.26322885,
        2.31105772,  1.99843906,  1.46912897,  1.53698319,  2.21166137,
        2.41131689,  2.88989157,  2.1295271 ,  3.51423149, 25.77165398,
       -3.10108965, 10.11547524, -0.17911783])
names = np.concatenate((np.array(["mu_{}_muN4258".format(n.decode('utf-8')) for n in sn_host]),np.array(["zp_w4258","b_w","m_v4258", "Z_w"])))
chimin = Minuit.from_array_func(chi2,x0, errordef = 1,error=0.001,name=names)

```

```python
print(chimin.migrad())
chimin.hesse()
print(chimin.errors)
```

```python
m0_4258 = chimin.np_values()[-2]
sigma_m0_4258 = chimin.np_errors()[-2]
```

```python
log_H0 = ((m0_4258-mu_N4258)+5*av+25)/5
sigma_log_H0 = np.sqrt(25*pw(sigma_av,2)+pw(sigma_mu_N4258,2)+pw(sigma_m0_4258,2))/5
H0 = pw(10,log_H0)
sigma_H0 = np.log(10)*pw(10,log_H0)*sigma_log_H0
print(f"logH0 = {log_H0} +/- {sigma_log_H0}")
print(f"H0 = {H0} +/- {sigma_H0}")
```

```python
mu_mu4258_fit = chimin.np_values()[:len(sn_host)]
zp = chimin.np_values()[len(sn_host)]
bw = chimin.np_values()[len(sn_host)+1]
zw = chimin.np_values()[len(sn_host)+3]
for host,i in zip(sn_host,np.arange(len(sn_host))):
    cephe_list = data_cephe[data_cephe['Field']== host]
    plt.figure()
    plt.title(host)
    plt.errorbar(cephe_list['Per'], cephe_list['F160W']-R*cephe_list['F555W-F814W'],yerr=cephe_list['sigma_tot'],fmt='.')
    p = np.linspace(np.min(cephe_list['Per'])-1,np.max(cephe_list['Per'])+1,1000)
    plt.plot(p,mu_mu4258_fit[i]+zp+bw*np.log10(p)+zw*np.mean(cephe_list['[O/H]']-12))
    plt.gca().invert_yaxis()
    plt.xscale('log')
    plt.xlabel('logP')
    plt.ylabel('mw')
plt.show()
```

```python
n_ceph_by_host = [len(data_cephe[data_cephe['Field']== host]) for host in sn_host]
n_ceph = np.sum(n_ceph_by_host)
n_ceph_cumsum = np.concatenate(([0],np.cumsum(n_ceph_by_host)))
n_sn = len(sn_host)
L_ceph = np.zeros((n_ceph,n_sn+4))
L_sn = np.zeros((n_sn,n_sn+4))
y_ceph=np.zeros(n_ceph)
y_sn = np.zeros(n_sn)
cov_ceph=np.zeros(n_ceph)
cov_sn=np.zeros(n_sn)
for i,host in enumerate(sn_host):
    selec_ceph = data_cephe[data_cephe['Field']== host]
    
    n_ceph_host = n_ceph_by_host[i]
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],i]=1
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-4]=1
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-3]=np.log10(selec_ceph['Per'])
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-1]=selec_ceph['[O/H]']-12
 
    L_sn[i,i] = 1 
    L_sn[i,-2] = 1
    
    y_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1]]= selec_ceph['F160W']-R*selec_ceph['F555W-F814W']
    y_sn[i] = data_sn[data_sn['Field']==host]['mag_5a'] - 5*av
    
    cov_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1]] = pw(selec_ceph['sigma_tot'],2)
    cov_sn[i] = pw(data_sn[data_sn['Field']==host]['sigma_mag_5a'],2)-25*pw(sigma_av,2)
    
selec_ceph = data_cephe[data_cephe['Field']== b'N4258']
n_4258 = len(data_cephe[data_cephe['Field']== b'N4258'])
L_4258 = np.zeros((n_4258,n_sn+4))
L_4258[:,-4] = 1
L_4258[:,-3] = np.log10(selec_ceph['Per'])
L_4258[:,-1] = selec_ceph['[O/H]']-12

y_4258 = selec_ceph['F160W']-R*selec_ceph['F555W-F814W']
cov_4258 = pw(selec_ceph['sigma_tot'],2)

L = np.concatenate((L_ceph,L_4258,L_sn),axis=0)
y = np.concatenate((y_ceph,y_4258,y_sn))
C=np.diag(np.concatenate((cov_ceph,cov_4258,cov_sn)))
```

```python
C_inv = np.linalg.inv(C)
cov_res = np.linalg.inv(L.T @ C_inv @ L)
res = cov_res @ L.T @ C_inv @ y
```

```python
print(res)
```

```python
log_H0 = (res[-2]-mu_N4258+5*av+25)/5
sigma_log_H0 = np.sqrt(25*pw(sigma_av,2)+pw(sigma_mu_N4258,2)+cov_res[-2,-2])/5
H0 = pw(10,log_H0)
sigma_H0 = np.log(10)*pw(10,log_H0)*sigma_log_H0
print(f"logH0 = {log_H0} +/- {sigma_log_H0}")
print(f"H0 = {H0} +/- {sigma_H0}")
```

```python
n_ceph_by_host = [len(data_cephe[data_cephe['Field']== host]) for host in sn_host]
n_ceph = np.sum(n_ceph_by_host)
n_ceph_cumsum = np.concatenate(([0],np.cumsum(n_ceph_by_host)))
n_sn = len(sn_host)
L_ceph = np.zeros((n_ceph,n_sn+5))
L_sn = np.zeros((n_sn,n_sn+5))
y_ceph=np.zeros(n_ceph)
y_sn = np.zeros(n_sn)
cov_ceph=np.zeros(n_ceph)
cov_sn=np.zeros(n_sn)
for i,host in enumerate(sn_host):
    selec_ceph = data_cephe[data_cephe['Field']== host]
    n_ceph_host = n_ceph_by_host[i]
    P_sup10 = selec_ceph['Per'] > 10
    P_inf10 = [not bool for bool in P_sup10]
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],i] = 1
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-5] = 1
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-4] = np.log10(selec_ceph['Per'])*P_sup10
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-3] = np.log10(selec_ceph['Per'])*P_inf10
    L_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1],-1]= selec_ceph['[O/H]']-12
 
    L_sn[i,i] = 1 
    L_sn[i,-2] = 1
    
    y_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1]]= selec_ceph['F160W']-R*selec_ceph['F555W-F814W']
    y_sn[i] = data_sn[data_sn['Field']== host]['mag_5a'] - 5*av
    
    cov_ceph[n_ceph_cumsum[i]:n_ceph_cumsum[i+1]] = pw(selec_ceph['sigma_tot'],2)
    cov_sn[i] = pw(data_sn[data_sn['Field']==host]['sigma_mag_5a'],2)-25*pw(sigma_av,2)
    
selec_ceph = data_cephe[data_cephe['Field']== b'N4258']
n_4258 = len(data_cephe[data_cephe['Field']== b'N4258'])
L_4258 = np.zeros((n_4258,n_sn+5))
P_sup10 = selec_cephe['Per'] > 10
P_inf10 = [not bool for bool in P_sup10]
L_4258[:,-5] = 1
L_4258[:,-4] = np.log10(selec_ceph['Per'])*P_sup10
L_4258[:,-3] = np.log10(selec_ceph['Per'])*P_inf10
L_4258[:,-1] = selec_ceph['[O/H]']-12

y_4258 = selec_ceph['F160W']-R*selec_ceph['F555W-F814W']
cov_4258 = pw(selec_ceph['sigma_tot'],2)

L = np.concatenate((L_ceph,L_4258,L_sn),axis=0)
y = np.concatenate((y_ceph,y_4258,y_sn))
C = np.diag(np.concatenate((cov_ceph,cov_4258,cov_sn)))
```

```python
C_inv = np.linalg.inv(C)
cov_res = np.linalg.inv(L.T @ C_inv @ L)
res = cov_res @ L.T @ C_inv @ y
```
